```python
gapminder_copy = gapminder.copy()
gapminder_copy.head()
```

```python
# drop rows with missing values
gapminder_copy = gapminder_copy.dropna()
print(len(gapminder))
print(len(gapminder_copy))
```

```python
# drop duplicate rows
gapminder_copy = gapminder_copy.drop_duplicates(keep = "first")
print(len(gapminder))
print(len(gapminder_copy))
```

```python
# converts values in 'year' & 'pop' columns to int type
gapminder_copy['year'] = gapminder_copy['year'].astype(int)
gapminder_copy['pop'] = gapminder_copy['pop'].astype(int)
print(gapminder.info())
print(gapminder_copy.info())
```

```python
# reset row index
gapminder_copy = gapminder_copy.reset_index(drop = True)
gapminder_copy
```

```python
# standardize format for string variables
gapminder_copy['region'] = gapminder_copy['region']. str.lstrip() # left side white space
gapminder_copy['region'] = gapminder_copy['region']. str.rstrip() # right side white space
gapminder_copy['region'] = gapminder_copy['region']. str.lower() # converts to lower case 
```

```python
# separate country and continent into separate columns
gapminder_copy['continent'] = gapminder_copy['region'].str.split('_', 1).str[0] # split by "_", 1 denotes column (vs row = 0) continent is 0th item
gapminder_copy['country'] = gapminder_copy['region'].str.split('_', 1).str[1] # split by "_", country is 1st item
gapminder_copy = gapminder_copy.drop('region', 1)
gapminder_copy.head()
```

```python
# change column name format
gapminder_copy.columns = gapminder_copy.columns.str.lower()
gapminder_copy = gapminder_copy.rename(columns = {'life exp':'lifeexp'}) # use dictionary to change column names
gapminder_copy.head()
```

```python
prb = pd.read_table("PRB_data.txt", sep = "\t")
prb_copy = prb.copy()
prb_copy.head()
```

```python
prb_copy['continent'] = prb_copy['region'].str.split('_', 1).str[0].str.lower()
prb_copy['country'] = prb_copy['region'].str.split('_', 1).str[1].str.lower()
prb_copy = prb_copy.drop('region', 1)
prb_copy.columns = prb_copy.columns.str.lower()
prb_copy = prb_copy.rename(columns = {'life exp':'lifeexp'})
```
